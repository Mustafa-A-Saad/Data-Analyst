```python
df["International plan"] = df["International plan"].map({"No":False,"Yes":True}) # don't run it twice 
# next time always make a copy of the original dataset
# map replaces values in a dict form { old_value : new_value }
```

```python
df.head()
```

```python
d = {"No":False , "Yes":True} # you can have the map dict as a variable 
df = df.replace({"Voice mail plan":d}) # this is the same as map
df.head()
```

```python
columns_to_show = ["Total day minutes","Total eve minutes","Total night minutes"] 
df.groupby(["Churn"])[columns_to_show].describe(percentiles=[]) # passing percentiles =[] removes interquatile percents (25%,etc..)
```

```python
columns_to_show = ["Total day minutes","Total eve minutes","Total night minutes"] 
df.groupby(["Churn"])[columns_to_show].agg([np.mean, np.std, np.min, np.max]) # passing a list to agg func
```

```python
pd.crosstab(df["Churn"],df["Voice mail plan"],normalize=True) # crosstab also called contingency table
```

```python
df.pivot_table(
values=["Total day calls","Total eve calls","Total night calls"],
index =["Area code"],
aggfunc=np.mean # or "mean"
)
```

```python
total_calls = (
    df["Total day calls"]
    + df["Total eve calls"]
    + df["Total night calls"]
    + df["Total intl calls"]
)

df.insert(loc=len(df.columns), column="Total calls", value=total_calls) # inserts a new column
```

```python
df.head()
```

```python
df["Total Charge"]=(
df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)  # inserts a new column , same as above without creating Series Instance
```

```python
# Deleting last 2 columns i just created using drop
df.drop(["Total Charge","Total calls"],axis=1,inplace=True)
# axis 1 to delete the column , not saying axis or 0 to delete row
df.head()
```
