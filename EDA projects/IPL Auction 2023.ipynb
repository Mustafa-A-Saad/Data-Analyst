```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Excels\\iplauction2023.csv")
df.head()
```

```python
sns.histplot(df['player style'])
plt.title("Histogram of player styles")
plt.xlabel("Player Style")
plt.ylabel("Count")
plt.xticks(rotation=90) # rotates the ticks on the x axis to 90 degree
plt.tight_layout()
plt.show()
```

```python
sns.histplot(df['franchise'])
plt.title("Players' Franchise")
plt.xlabel("Francise")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
columns_to_plot = ['nationality','status']

fig , axes = plt.subplots(nrows=1,ncols=2,figsize=(8,6),edgecolor="black")

#fig = figure itself   # axes = is each indiviual subplot in this case we have 2 subplots so 2 axes

for i,column in enumerate(columns_to_plot):
    sns.histplot(df[column],ax=axes[i])
    axes[i].set_title(f"Distribution of {column.capitalize()}")
    axes[i].set_xlabel(column.capitalize())
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis="x",rotation=90)

plt.tight_layout()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['base price (in lacs)'],df['final price (in lacs)'],color='blue',alpha=0.5)
plt.title("Scatter Plot of Base Price vs Final Price")
plt.xlabel("Base Price (in lacs)")
plt.ylabel("Final Price (in lacs)")
```
