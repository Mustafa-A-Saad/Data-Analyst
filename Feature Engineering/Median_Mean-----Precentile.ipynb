```python
min_threshold,max_threshold = df['price'].quantile([0.01,0.999])
min_threshold,max_threshold
```

```python
df[df.price<min_threshold].head()
```

```python
df2 = df[ (df.price>min_threshold)&(df.price<max_threshold)]
df2.shape
```
