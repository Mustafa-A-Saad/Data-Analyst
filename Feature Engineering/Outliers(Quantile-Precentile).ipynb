```python
import pandas as pd

df = pd.read_csv("Excels\heights.csv")
df.head()
```

```python
max_threshold = df['height'].quantile(0.95) # 95% percent rank
```

```python
df[df['height']>max_threshold]
```

```python
min_threshold = df['height'].quantile(0.05) # 5% percent rank
min_threshold
```

```python
df[df['height']<min_threshold]
```

```python
df = df[(df['height'] < max_threshold) & (df['height'] > min_threshold)]
```
