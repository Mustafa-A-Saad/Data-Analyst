```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Excels\\heights2.csv")
df.head()
```

```python
plt.hist(df['height'],bins=20,edgecolor='black',rwidth=0.8)
plt.ylabel("Count")
plt.show()
```

```python
from scipy.stats import norm

plt.hist(df['height'],bins=20,edgecolor='black',rwidth=0.8,density=True)
plt.ylabel("Count")
plt.xlabel("Height (Inches)")

rng = np.arange(df['height'].min(),df['height'].max(),0.1)
plt.plot(rng , norm.pdf(rng,df['height'].mean(),df['height'].std()))
```

```python
upper_limit = df['height'].mean()+3*df['height'].std()
upper_limit
```

```python
lower_limit = df['height'].mean()-3*df['height'].std()
lower_limit
```

```python
df[ (df['height']>upper_limit) | (df['height']<lower_limit)]
```

```python
df = df[ (df['height'] < upper_limit) & (df['height'] > lower_limit)] # uinsg Std devs ( Emprical Rule )
```

```python
df['z-score'] = (df['height']-df['height'].mean())/df['height'].std()
df.head()
```

```python
df[ (df['z-score'] > 3) | (df['z-score']<-3) ] # no outliers
```
